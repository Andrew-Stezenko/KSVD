```python
import turtle

class LSystem2D:
    def __init__(self, t, size, n):
        self.size = size
        self.n = n
        self.t = t
        self.t.speed(0)
        self.t.hideturtle()

    def draw_square(self, size):
        for _ in range(4):
            self.t.forward(size)
            self.t.left(90)

    def draw_carpet(self, size, n):
        if n == 0:
            self.draw_square(size)
        else:
            step = size / 3
            for i in range(3):
                for j in range(3):
                    if i == 1 and j == 1:
                        pass
                    else:
                        self.t.penup()
                        self.t.goto(self.t.xcor() + i * step, self.t.ycor() - j * step)
                        self.t.pendown()
                        self.draw_carpet(step, n - 1)
                        self.t.penup()
                        self.t.goto(self.t.xcor() - i * step, self.t.ycor() + j * step)
                        self.t.pendown()

    def draw_sierpinski_carpet(self):
        initial_pos = self.t.position()
        initial_heading = self.t.heading()
        self.t.penup()
        self.t.goto(-self.size / 2, self.size / 2)
        self.t.pendown()
        self.draw_carpet(self.size, self.n)
        self.t.penup()
        self.t.goto(initial_pos)
        self.t.setheading(initial_heading)


width = 800
height = 800
screen = turtle.Screen()
screen.setup(width, height, 0, 0)

t = turtle.Turtle()
t.speed(0)
t.penup()
t.pendown()

carpet = LSystem2D(t, 600, 3)
carpet.draw_sierpinski_carpet()

turtle.done()
```
