```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('spotify-2023.csv', delimiter=',')
```

```python
data.head()
```

Сортуємо пісні за кількістю прослуховувань

```python
sort_data = data.sort_values(by='streams', ascending=False)
```

```python
top_10 = sort_data.head(10)
```

```python
sort_data.head(10)
```

1. Побудувати горизонтальну стовпчикову діаграму, на якій відображаються 10 
найпопулярніших пісень на Spotify у 2023 році за кількістю 
прослуховувань

```python
plt.figure(figsize=(15, 8))
plt.barh(top_10['track_name'], top_10['streams'], color='blue')
plt.xlabel('Кількість прослуховувань')
plt.ylabel('Назва пісні')
plt.title('Топ 10 найпопулярніших пісень на Spotify у 2023 році')
plt.gca().invert_yaxis()
plt.show()
```

Побудувати кругову діаграму для тих самих даних

```python
plt.figure(figsize=(8, 8))
plt.pie(top_10['streams'], labels=top_10['track_name'], autopct='%1.1f%%', startangle=140)
plt.title('Топ 10 найпопулярніших пісень на Spotify у 2023 році')
plt.axis('equal')
plt.show()
```

2. Створіть кругову діаграму або секторну діаграму, яка відображає розподіл
за danceability серед найпопулярніших пісень у 2023 році.

```python
plt.figure(figsize=(8, 8))
plt.pie(top_10['danceability_%'], labels=top_10['track_name'], autopct='%1.1f%%', startangle=140)
plt.title('Розподіл за показником danceability серед найпопулярніших пісень на Spotify у 2023 році')
plt.axis('equal')
plt.show()
```

4 Створіть стовпчикову діаграму, на якій відображаються 10 найпопулярніших 
артистів на Spotify у 2023 році за кількістю слухачів.

```python
plt.figure(figsize=(10, 6))
plt.bar(top_10['artist(s)_name'], top_10['streams'], color='blue')
plt.xlabel('Артист')
plt.ylabel('Кількість слухачів')
plt.title('10 найпопулярніших артистів на Spotify у 2023 році')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

5. Побудуйте горизонтальну стовпчикову діаграму, яка порівнює кількість 
прослуховувань трьох конкретних треків у 2023 році

```python
track1 = 'Love Again'
track2 = 'TQG'
track3 = 'HAPPY'

track1_data = data[(data['track_name'] == track1)]
track2_data = data[(data['track_name'] == track2)]
track3_data = data[(data['track_name'] == track3)]

plt.figure(figsize=(10, 6))
plt.barh([track1, track2, track3], [track1_data['streams'].sum(), track2_data['streams'].sum(), track3_data['streams'].sum()], color=['blue', 'red', 'green'])
plt.xlabel('Кількість прослуховувань')
plt.ylabel('Назва треку')
plt.title('Порівняння кількості прослуховувань трьох конкретних треків у 2023 році')
plt.show()
```
